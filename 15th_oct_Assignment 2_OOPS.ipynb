```python
#3. How do you define a constructor in a Python class? Provide an example.

#In Python, a constructor is defined within a class using a special method called __init__. 
#The purpose of the constructor is to initialize the #attributes of an object when an instance of the class is created. Here's an example:

class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        self.is_engine_on = False

    def start_engine(self):
        self.is_engine_on = True
        print("Engine started.")

    def stop_engine(self):
        self.is_engine_on = False
        print("Engine stopped.")

# Creating an instance of the Car class
my_car = Car(make="Toyota", model="Camry", year=2022)

# Accessing attributes
print(f"My car is a {my_car.year} {my_car.make} {my_car.model}.")
print(f"Is the engine on? {'Yes' if my_car.is_engine_on else 'No'}")

# Starting the engine
my_car.start_engine()
print(f"Is the engine on? {'Yes' if my_car.is_engine_on else 'No'}")

# Stopping the engine
my_car.stop_engine()
print(f"Is the engine on? {'Yes' if my_car.is_engine_on else 'No'}")

```

```python
#10. How can you have multiple constructors in a Python class? Explain with an example.

#In Python, a class can't have multiple constructors in the same way some other languages 
#might support method overloading. However, you can achieve similar functionality using default
#parameter values and class methods. By providing default values to parameters or creating class 
#methods that act as alternative constructors, you can simulate the effect of having multiple constructors.

#Here's an example demonstrating how you can have multiple "constructors" in a Python class:

class MyClass:
    def __init__(self, attribute1, attribute2):
        self.attribute1 = attribute1
        self.attribute2 = attribute2

    @classmethod
    def from_string(cls, input_string):
        # Parse the input string and extract values
        attribute1, attribute2 = input_string.split(',')
        
        # Create and return an instance using the extracted values
        return cls(attribute1, attribute2)

    @classmethod
    def default_constructor(cls):
        # Create and return an instance with default values
        return cls(attribute1="default_value1", attribute2="default_value2")

# Using the primary constructor
instance1 = MyClass("value1", "value2")
print(f"Instance 1: {instance1.attribute1}, {instance1.attribute2}")

# Using an alternative constructor (from_string)
instance2 = MyClass.from_string("custom_value1,custom_value2")
print(f"Instance 2: {instance2.attribute1}, {instance2.attribute2}")

# Using another alternative constructor (default_constructor)
instance3 = MyClass.default_constructor()
print(f"Instance 3: {instance3.attribute1}, {instance3.attribute2}")


#In this example:

#The primary constructor (__init__) takes two parameters (attribute1 and attribute2).
#There are two class methods: from_string and default_constructor.

#from_string takes a string as input, parses it, and creates an instance using the extracted values.

#default_constructor creates an instance with default values.

#By using class methods, you provide alternative ways to construct instances of the class. 
#You can then choose the appropriate method based on the input or requirements. Note that the 
#use of @classmethod decorators is essential for these methods to work with the class rather than an instance.

#This approach allows you to create a more flexible and expressive interface for creating instances of your class,
#giving users different options depending on their needs.
```

```python
#11. What is method overloading, and how is it related to constructors in Python?

# Method overloading is a concept in object-oriented programming where a class can have multiple methods with the
# same name but different parameter lists. The idea is to provide flexibility in calling a method by allowing it to 
# behave differently based on the number or types of parameters passed to it.

# In Python, method overloading is not supported in the traditional sense as it is in some other programming languages
# (like Java or C++). However, Python provides a flexible approach to achieve similar functionality using default parameter values and variable-length argument lists. This means that you can define a single method that can handle multiple parameter variations.

# When it comes to constructors in Python, method overloading is relevant because constructors are essentially special
# methods. While Python doesn't support method overloading in the strict sense of the term, you can achieve similar effects by defining multiple constructors using default values or variable-length arguments.

# Here's an example demonstrating a form of method overloading in a Python class:

class MyClass:
    def __init__(self, attribute1, attribute2=None):
        if attribute2 is None:
            # Handle the case where attribute2 is not provided
            self.attribute1 = attribute1
            self.attribute2 = "default_value"
        else:
            # Handle the case where both attribute1 and attribute2 are provided
            self.attribute1 = attribute1
            self.attribute2 = attribute2

# Using the constructor with one argument
instance1 = MyClass("value1")
print(f"Instance 1: {instance1.attribute1}, {instance1.attribute2}")

# Using the constructor with two arguments
instance2 = MyClass("value1", "value2")
print(f"Instance 2: {instance2.attribute1}, {instance2.attribute2}")


# In this example:

# The __init__ method is a constructor that takes two parameters (attribute1 and attribute2).
# The default value of None for attribute2 allows you to call the constructor with either one or two arguments.
# Inside the constructor, a condition checks whether attribute2 is None to determine if it was provided during the 
# object creation.
# While Python doesn't have method overloading in the traditional sense, this approach allows you to create flexible 
# constructors that can handle different parameter variations. The choice of which variation to use depends on the 
# arguments provided during object creation.
```

```python
#12. Explain the use of the `super()` function in Python constructors. Provide an example.

# In Python, the super() function is used to call a method from a parent class. It is often 
# used within a subclass's method (including constructors) to invoke the method of the parent 
# class, allowing you to extend or override its behavior.

# In the context of constructors, super().__init__() is commonly used to call the constructor of 
# the parent class. This is particularly useful when you're working with inheritance and want to 
# initialize attributes in both the subclass and the superclass.

# Here's an example to illustrate the use of super() in a Python constructor:

class ParentClass:
    def __init__(self, parent_attribute):
        self.parent_attribute = parent_attribute

    def display(self):
        print(f"Parent Attribute: {self.parent_attribute}")

class ChildClass(ParentClass):
    def __init__(self, parent_attribute, child_attribute):
        # Call the constructor of the parent class using super()
        super().__init__(parent_attribute)
        
        # Initialize attributes specific to the child class
        self.child_attribute = child_attribute

    def display(self):
        # Call the display method of the parent class using super()
        super().display()
        print(f"Child Attribute: {self.child_attribute}")

# Create an instance of the child class
child_instance = ChildClass(parent_attribute="Parent Value", child_attribute="Child Value")

# Call the display method of the child class
child_instance.display()
```

```python
#16. How do you prevent a class from having multiple instances by using constructors in Python? Provide an example.

# To prevent a class from having multiple instances (implementing a Singleton pattern) in Python, you 
# can use a class variable to track whether an instance has already been created. If an instance 
# already exists, subsequent attempts to create new instances can return the existing instance rather
# than creating a new one. This ensures that only one instance of the class is created during the program's execution.

class SingletonClass:
    _instance = None

    def __new__(cls, *args, **kwargs):
        if cls._instance is None:
            cls._instance = super(SingletonClass, cls).__new__(cls)
            # Initialization code here
        return cls._instance

    def __init__(self, data):
        # Initialization code here
        if not hasattr(self, '_initialized'):
            self.data = data
            self._initialized = True

# Creating instances of SingletonClass
singleton_instance1 = SingletonClass(data="Instance 1 Data")
singleton_instance2 = SingletonClass(data="Instance 2 Data")

# Checking if both instances refer to the same object
print(f"Is instance1 the same as instance2? {singleton_instance1 is singleton_instance2}")

# Accessing and printing data from both instances
print(f"Data in instance 1: {singleton_instance1.data}")
print(f"Data in instance 2: {singleton_instance2.data}")

# Modifying data in one instance
singleton_instance1.data = "Modified Data"

# Checking if the modification is reflected in the other instance
print(f"Data in instance 1 after modification: {singleton_instance1.data}")
print(f"Data in instance 2 after modification: {singleton_instance2.data}")

# In this example:

# The __new__ method ensures that only one instance is created by checking if _instance is None. If it is None, a new instance is created; otherwise, the existing instance is returned.
# The __init__ method is used to initialize instance variables, and it checks if the instance has been initialized before using _initialized.
# This implementation ensures that only one instance of the SingletonClass is created, and subsequent attempts to create new instances return the existing instance.
```

```python
#19. Explain the use of constructor chaining in Python. Provide a practical example.

# Constructor chaining in Python refers to the concept of one constructor calling another 
# constructor within the same class or in its superclass. This allows you to reuse code
# and avoid redundancy when setting up object attributes or performing certain initializations.
# The super() function is often used for constructor chaining to call the constructor of the superclass.

# Here's an example to illustrate constructor chaining:

class Animal:
    def __init__(self, species):
        self.species = species
        print(f"Animal constructor called. Species: {self.species}")

class Mammal(Animal):
    def __init__(self, species, sound):
        super().__init__(species)  # Calling the constructor of the superclass
        self.sound = sound
        print(f"Mammal constructor called. Sound: {self.sound}")

class Dog(Mammal):
    def __init__(self, breed, sound):
        super().__init__(species="Dog", sound=sound)  # Chaining to the Mammal constructor
        self.breed = breed
        print(f"Dog constructor called. Breed: {self.breed}")

# Creating an instance of Dog
my_dog = Dog(breed="Golden Retriever", sound="Woof")
```

```python
# 1. What is inheritance in Python? Explain its significance in object-oriented programming.

# Inheritance is a fundamental concept in object-oriented programming (OOP) that allows a new class, known as the
# subclass or derived class, to inherit attributes and behaviors (methods) from an existing class, known as the
# superclass or base class. Inheritance enables code reuse and establishes relationships between classes, creating 
# a hierarchy of classes with shared characteristics.

# syntax:
class Superclass:
    # Superclass attributes and methods

class Subclass(Superclass):
    # Subclass inherits from Superclass
    # Subclass attributes and methods can be added or overridden

# Types of Inheritance:

# Single Inheritance: A subclass inherits from only one superclass.
# Multiple Inheritance: A subclass inherits from more than one superclass.
# Multilevel Inheritance: A subclass inherits from another subclass, creating a hierarchy.

# Significance in OOP:

# Code Reusability: Inheritance allows you to reuse code from existing classes, reducing redundancy and promoting a 
# modular and maintainable codebase.
# Polymorphism: Inherited methods can be overridden in the subclass to provide specialized behavior. This supports 
# polymorphism, where objects of different classes can be treated as objects of a common base class.
# Hierarchy and Organization: Inheritance helps create a hierarchy of classes that reflects the relationships between 
# entities in the problem domain. This hierarchy enhances the organization of code and improves its readability.



#example:

class Animal:
    def __init__(self, species):
        self.species = species

    def make_sound(self):
        print("Some generic animal sound")

class Dog(Animal):
    def __init__(self, breed):
        super().__init__(species="Dog")
        self.breed = breed

    def make_sound(self):
        print("Woof! Woof!")

my_dog = Dog(breed="Golden Retriever")
my_dog.make_sound()  # Output: Woof! Woof!
```

```python
# 2. Differentiate between single inheritance and multiple inheritance in Python. Provide examples for each.

# Single Inheritance:

# In single inheritance, a class inherits from only one superclass. This is the simplest form of inheritance.

# Example of Single Inheritance:

class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

# Creating an instance of Dog
my_dog = Dog()

# Accessing methods from both Animal and Dog
my_dog.speak()  # Output: Animal speaks
my_dog.bark()   # Output: Dog barks


# In this example, the Dog class inherits from the Animal class. It can access both the speak method from Animal 
# and its own bark method.




# Multiple Inheritance:

# In multiple inheritance, a class can inherit from more than one superclass. This allows the subclass to inherit
# attributes and methods from multiple classes.

# Example of Multiple Inheritance:

class Flying:
    def fly(self):
        print("Can fly")

class Swimming:
    def swim(self):
        print("Can swim")

class FlyingFish(Flying, Swimming):
    pass

# Creating an instance of FlyingFish
my_fish = FlyingFish()

# Accessing methods from both Flying and Swimming
my_fish.fly()   # Output: Can fly
my_fish.swim()  # Output: Can swim


# In this example, the FlyingFish class inherits from both the Flying and Swimming classes, allowing it to access 
# methods from both superclasses.

# Key Differences:

# Number of Superclasses:

# Single Inheritance: Inherits from only one superclass.
# Multiple Inheritance: Inherits from more than one superclass.

# Simplicity:

# Single Inheritance: Simpler and easier to understand.
# Multiple Inheritance: Can be more complex, and the class hierarchy needs careful design.

# Example Use Cases:

# Single Inheritance: Used when there is a clear, linear relationship between classes.
# Multiple Inheritance: Used when a class needs to combine features from multiple sources.

# Ambiguity:

# Single Inheritance: No issue of method ambiguity.
# Multiple Inheritance: Potential for method ambiguity if multiple superclasses have methods with the same name.
```

```python
# 18. What is the "diamond problem" in multiple inheritance, and how does Python address it?

# The "diamond problem" is a term used in the context of multiple inheritance in object-oriented programming languages. 
# It refers to a potential ambiguity and complexity that arises when a class inherits from two classes that have a common
# ancestor. The diamond problem occurs when there is a chain of inheritance such that a class (let's call it D) inherits
# from two classes (B and C), both of which inherit from a common base class (A). In other words, the inheritance hierarchy 
# forms a diamond shape.

#       A
#      / \
#     B   C
#      \ /
#       D

# In this scenario:

# Class D inherits from both classes B and C.
# Both B and C inherit from the common base class A.
# The problem arises when there are conflicting methods or attributes in class A that are inherited by both B and C. 
# If class D does not provide its own implementation of these conflicting methods or attributes, it becomes unclear 
# which one to use, leading to ambiguity and potential issues.

# Python addresses the diamond problem using a mechanism called "C3 linearization" or the C3 superclass linearization
# algorithm. This algorithm ensures a consistent and predictable order for method resolution in the presence of multiple
# inheritance.

# When a method is called on an object in a class hierarchy, Python follows the C3 linearization order to determine the 
# order in which the classes are considered for method resolution. The order is based on the depth-first, left-to-right 
# traversal of the inheritance hierarchy.


# Example:

class A:
    def hello(self):
        print("Hello from class A")

class B(A):
    def hello(self):
        print("Hello from class B")
        super().hello()

class C(A):
    def hello(self):
        print("Hello from class C")
        super().hello()

class D(B, C):
    pass

# Example usage
obj_d = D()
obj_d.hello()

Output:

Hello from class B
Hello from class C
Hello from class A


# Explanation:

# D inherits from B, so the hello method in class B is called first.
# Inside the hello method of class B, super().hello() calls the hello method of the next class in the linearization order, which is C.
# The hello method in class C is called, and it, in turn, calls the hello method of class A using super().hello().
# Finally, the hello method of class A is called.
```

```python
# 4. Discuss the difference between public, private, and protected access modifiers in Python.

# In Python, access modifiers are used to control the visibility of class members (attributes and methods) from outside
# the class. The three main access modifiers are public, private, and protected. Here's an overview of each:

# 1. Public Access Modifier:
# Syntax: No special syntax is used. Members are public by default.
# Naming Convention: Members with public access are typically named without any leading underscores or double underscores.
# Visibility: Public members can be accessed from outside the class.
# Example:

class MyClass:
    def public_method(self):
        return "This is a public method"
        
# 2. Protected Access Modifier:
# Syntax: Members with protected access are named with a single leading underscore (e.g., _variable).
# Visibility: Protected members can be accessed from within the class and its subclasses.
# Note: Python does not enforce strict protection, and it relies on convention and programmer discipline.
# Example:


class MyClass:
    def __init__(self):
        self._protected_variable = 42

    def _protected_method(self):
        return "This is a protected method"
        
# 3. Private Access Modifier:
# Syntax: Members with private access are named with a double leading underscore (e.g., __variable).
# Visibility: Private members are not accessible directly from outside the class. However, Python uses name mangling, and it is still possible to access them using a modified name (e.g., _ClassName__variable).
# Note: Private members are intended to be used within the class only, and direct access from outside the class is discouraged.
# Example:

class MyClass:
    def __init__(self):
        self.__private_variable = 42

    def __private_method(self):
        return "This is a private method"
```

```python
# 6. Explain the purpose of getter and setter methods in encapsulation. Provide examples.

# Getter and setter methods play a crucial role in encapsulation by providing controlled access to the
# attributes of a class. They allow you to retrieve (get) and modify (set) the values of private or
# protected attributes while enforcing rules or validation logic if needed. This helps in maintaining 
# the integrity of the object's state and adheres to the principles of information hiding.

# Purpose of Getter Methods:
# Getters are used to retrieve the values of private or protected attributes:
# They provide a way to access the internal state of an object from outside the class.
# Getters are typically named with the prefix "get" followed by the attribute name.

# Example:

class Student:
    def __init__(self, name, age):
        self._name = name  # Protected attribute
        self._age = age    # Protected attribute

    def get_name(self):
        return self._name

    def get_age(self):
        return self._age

# Example usage
student = Student("John", 20)
print(student.get_name())  # Output: John
print(student.get_age())   # Output: 20


# Purpose of Setter Methods:
# Setters are used to modify the values of private or protected attributes:
# They provide a controlled way to update the internal state of an object.
# Setters are typically named with the prefix "set" followed by the attribute name.
# They often include validation logic to ensure that the new value meets certain criteria.

# Example:

class Car:
    def __init__(self, make, model):
        self._make = make  # Protected attribute
        self._model = model  # Protected attribute

    def set_make(self, make):
        if make.isalpha():  # Example validation logic
            self._make = make
        else:
            print("Invalid make value. Please provide a valid string.")

    def set_model(self, model):
        if isinstance(model, str):  # Example validation logic
            self._model = model
        else:
            print("Invalid model value. Please provide a valid string.")

# Example usage
car = Car("Toyota", "Camry")
car.set_make("Honda")  # Valid update
car.set_model(123)     # Invalid update, prints an error message


# In these examples, the getter methods (get_name, get_age) allow external code to retrieve the values of 
# protected attributes _name and _age. The setter methods (set_make, set_model) provide a controlled way
# to modify the values of protected attributes _make and _model while incorporating validation logic.

# Using getters and setters instead of direct attribute access ensures that any necessary logic, validation, or side effects 
# are applied consistently when reading or modifying the internal state of an object. This is a fundamental aspect of 
# encapsulation, promoting better code organization and maintainability.
```

```python
# 7. What is name mangling in Python, and how does it affect encapsulation?

# Name mangling is a mechanism in Python that transforms the names of attributes in a class to make them less accessible
# from outside the class. It is applied by adding a prefix to the names of attributes that are marked as private with 
# a double leading underscore (__). This prefix is based on the name of the class itself, making it more challenging
# (though not impossible) for external code to access or modify these attributes directly.

# Purpose of Name Mangling:

# Avoiding Name Clashes:

# Name mangling helps prevent accidental name clashes between attributes in different classes. Each class gets a unique 
# prefix based on its name, reducing the likelihood of unintentional conflicts.

# Limited Encapsulation:

# While name mangling is a form of encapsulation, it is not meant to provide strong security or access control. It serves
# more as a convention to discourage external access rather than a strict enforcement mechanism.

# Example:

class MyClass:
    def __init__(self):
        self.__private_variable = 42

    def __private_method(self):
        return "This is a private method"

# Example usage
obj = MyClass()

# Accessing private attribute using name mangling
print(obj._MyClass__private_variable)  # Output: 42

# Accessing private method using name mangling
print(obj._MyClass__private_method())  # Output: This is a private method

# In this example, the attribute __private_variable and the method __private_method are subject to name mangling.
# Accessing them from outside the class requires using the modified names with the class 
# prefix (_MyClass__private_variable and _MyClass__private_method).

# Impact on Encapsulation:
# Limited Access Control:

# Name mangling does not provide strict access control. It is a convention rather than a security feature, and it can
# be bypassed by determined users.

# Maintaining Intent of Encapsulation:

# While name mangling makes it more challenging for external code to access private attributes, it is essential to 
# adhere to the intent of encapsulation and use the provided interfaces (getter and setter methods) to interact with
# the object's state.

# Code Readability:

# The use of name mangling may impact code readability, and it is generally recommended to follow good practices in 
# naming conventions and to use encapsulation judiciously.
```

```python
# 8. Create a Python class called `BankAccount` with private attributes for the account balance (`__balance`) and account number (`__account_number`). Provide methods for depositing and withdrawing money.

class BankAccount:
    def __init__(self, balance=0, account_number=""):
        self.__balance = balance
        self.__account_number = account_number

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            return self.__balance  # Return the updated balance
        else:
            return "Enter amount greater than 0"

    def withdraw(self, amount):
        if amount > 0 and self.__balance >= amount:
            self.__balance -= amount
            return self.__balance  # Return the updated balance
        else:
            return "Withdraw not successful"

    def show_balance(self):
        return self.__balance

# Example usage
acc = BankAccount(100, "10201")
print("Initial Balance:", acc.show_balance())

updated_balance = acc.deposit(1000)
print("After Deposit:", updated_balance)

updated_balance = acc.withdraw(10)
print("After Withdrawal:", updated_balance)



        
```

```python
# 11. Create a Python class hierarchy for a school system, including classes for students, teachers, and courses, and implement encapsulation principles to protect sensitive information.

class Person:
    def __init__(self, name, age):
        self._name = name  # Protected attribute
        self._age = age    # Protected attribute

    def display_info(self):
        print(f"Name: {self._name}")
        print(f"Age: {self._age}")


class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)
        self.__student_id = student_id  # Private attribute

    def display_info(self):
        super().display_info()
        print(f"Student ID: {self.__student_id}")

    def get_student_id(self):
        return self.__student_id

    def set_student_id(self, student_id):
        self.__student_id = student_id


class Teacher(Person):
    def __init__(self, name, age, employee_id):
        super().__init__(name, age)
        self.__employee_id = employee_id  # Private attribute

    def display_info(self):
        super().display_info()
        print(f"Employee ID: {self.__employee_id}")

    def get_employee_id(self):
        return self.__employee_id

    def set_employee_id(self, employee_id):
        self.__employee_id = employee_id


class Course:
    def __init__(self, course_code, course_name):
        self._course_code = course_code  # Protected attribute
        self._course_name = course_name  # Protected attribute
        self.__students = []  # Private attribute
        self.__teacher = None  # Private attribute

    def assign_teacher(self, teacher):
        if isinstance(teacher, Teacher):
            self.__teacher = teacher

    def enroll_student(self, student):
        if isinstance(student, Student):
            self.__students.append(student)

    def display_info(self):
        print(f"Course Code: {self._course_code}")
        print(f"Course Name: {self._course_name}")
        if self.__teacher:
            print("Assigned Teacher:")
            self.__teacher.display_info()
        if self.__students:
            print("Enrolled Students:")
            for student in self.__students:
                student.display_info()


# Example usage
student1 = Student("Alice", 18, "S001")
teacher1 = Teacher("Mr. Smith", 35, "T001")
course1 = Course("C101", "Mathematics")

course1.assign_teacher(teacher1)
course1.enroll_student(student1)

course1.display_info()
```

```python
# 12. Explain the concept of property decorators in Python and how they relate to encapsulation.

# In Python, property decorators are a way to encapsulate attributes of a class by providing 
# a controlled interface for accessing and modifying them. Property decorators allow you to define
# getter, setter, and deleter methods for an attribute, giving you more control over how attribute 
# access and modifications are handled. This contributes to encapsulation by allowing you to hide
# the implementation details of the attribute and provide controlled access to it.

#example:

class Person:
    def __init__(self, name):
        self._name = name  # Protected attribute

    @property              #It allows you to access the attribute as if it were a regular attribute
    def name(self):
        return self._name

    @name.setter           #It allows you to control how the attribute is modified, validating the new value or performing additional actions.
    def name(self, value):
        if isinstance(value, str):
            self._name = value
        else:
            print("Invalid value. Please provide a string.")

    @name.deleter #It allows you to perform cleanup or additional actions when the attribute is deleted.
    def name(self):
        print(f"Deleting {self._name}")
        del self._name

# Example usage
person = Person("John")
print(person.name)      # Accessing the attribute using the getter method
person.name = "Alice"   # Modifying the attribute using the setter method
del person.name         # Deleting the attribute using the deleter method



# Property decorators help encapsulate the implementation details of an attribute by providing 
# controlled access through getter, setter, and deleter methods.
# They allow you to add logic, validation, and additional actions without exposing
# the internal representation of the attribute.
```

```python
# 7. Discuss the use of abstract methods and classes in achieving polymorphism in Python. Provide an example using the `abc` module.

# Abstract classes and abstract methods play a crucial role in achieving polymorphism in Python. 
# Abstract classes provide a blueprint for other classes and can define abstract methods that must
# be implemented by any concrete (non-abstract) subclasses. The abc module in Python is commonly used
# to create abstract classes and abstract methods.

# Here's an example using the abc module to demonstrate abstract classes and methods for achieving polymorphism:

from abc import ABC, abstractmethod

# Define an abstract class with an abstract method
class Shape(ABC):
    @abstractmethod
    def calculate_area(self):
        pass

# Concrete subclasses implementing the abstract class
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def calculate_area(self):
        return 3.14 * self.radius**2

class Square(Shape):
    def __init__(self, side):
        self.side = side

    def calculate_area(self):
        return self.side**2

class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def calculate_area(self):
        return 0.5 * self.base * self.height

# Function demonstrating polymorphism using the abstract class
def print_area(shape):
    print(f"Area is: {shape.calculate_area()}")

# Creating objects of different shapes
circle = Circle(5)
square = Square(4)
triangle = Triangle(3, 6)

# Using polymorphism through the abstract class
print_area(circle)    # Outputs: Area is: 78.5
print_area(square)    # Outputs: Area is: 16
print_area(triangle)  # Outputs: Area is: 9.0

# In this example:

# The Shape class is an abstract class that defines an abstract method calculate_area(). Any concrete
# subclass of Shape must implement this method.

# The Circle, Square, and Triangle classes are concrete subclasses that inherit from Shape and provide 
# their own implementations of the calculate_area() method.

# The print_area function demonstrates polymorphism by accepting any object of a class derived from Shape
# and calling its calculate_area() method.

# By using abstract classes and methods, you ensure that all subclasses adhere to a common interface,
# allowing polymorphic behavior while providing a clear structure for your class hierarchy.
```

```python
# 16. Create a Python class for a computer system, demonstrating abstraction by defining common methods (e.g., `power_on()`, `shutdown()`) in an abstract base class.

from abc import ABC, abstractmethod

# Abstract Base Class for Computer System
class ComputerSystem(ABC):
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
        self.powered_on = False

    @abstractmethod
    def power_on(self):
        pass

    @abstractmethod
    def shutdown(self):
        pass

# Concrete Classes for Specific Computer Systems
class Laptop(ComputerSystem):
    def __init__(self, brand, model, battery_life):
        super().__init__(brand, model)
        self.battery_life = battery_life

    def power_on(self):
        self.powered_on = True
        print(f"{self.brand} {self.model} Laptop is powered on.")

    def shutdown(self):
        self.powered_on = False
        print(f"{self.brand} {self.model} Laptop is shutting down.")

class Desktop(ComputerSystem):
    def __init__(self, brand, model, monitor_type):
        super().__init__(brand, model)
        self.monitor_type = monitor_type

    def power_on(self):
        self.powered_on = True
        print(f"{self.brand} {self.model} Desktop is powered on with {self.monitor_type} monitor.")

    def shutdown(self):
        self.powered_on = False
        print(f"{self.brand} {self.model} Desktop is shutting down.")

# Example Usage
laptop = Laptop("Dell", "XPS 13", "8 hours")
desktop = Desktop("HP", "Pavilion", "LED")

laptop.power_on()
desktop.power_on()

laptop.shutdown()
desktop.shutdown()
```

```python
# 19. Create a Python class for a library system, implementing abstraction by defining common methods (e.g., `add_book()`, `borrow_book()`) in an abstract base class.

from abc import ABC, abstractmethod

# Abstract Base Class for Library System
class LibrarySystem(ABC):
    def __init__(self, name):
        self.name = name
        self.books = {}  # Dictionary to store books and their availability

    @abstractmethod
    def add_book(self, book_title, author):
        pass

    @abstractmethod
    def borrow_book(self, book_title):
        pass

    @abstractmethod
    def return_book(self, book_title):
        pass

    def display_books(self):
        print(f"\nBooks available in {self.name} library:")
        for title, availability in self.books.items():
            status = "Available" if availability else "Borrowed"
            print(f"{title} by {self.books[title]['author']} ({status})")

# Concrete Class for Library Implementation
class PublicLibrary(LibrarySystem):
    def add_book(self, book_title, author):
        if book_title not in self.books:
            self.books[book_title] = {'author': author, 'available': True}
            print(f"Book '{book_title}' by {author} added to the library.")
        else:
            print(f"Book '{book_title}' already exists in the library.")

    def borrow_book(self, book_title):
        if book_title in self.books and self.books[book_title]['available']:
            self.books[book_title]['available'] = False
            print(f"Book '{book_title}' borrowed successfully.")
        elif book_title in self.books and not self.books[book_title]['available']:
            print(f"Book '{book_title}' is currently not available.")
        else:
            print(f"Book '{book_title}' does not exist in the library.")

    def return_book(self, book_title):
        if book_title in self.books and not self.books[book_title]['available']:
            self.books[book_title]['available'] = True
            print(f"Book '{book_title}' returned successfully.")
        elif book_title in self.books and self.books[book_title]['available']:
            print(f"Book '{book_title}' is already available in the library.")
        else:
            print(f"Book '{book_title}' does not exist in the library.")

# Example Usage
public_library = PublicLibrary("City Library")

public_library.add_book("The Great Gatsby", "F. Scott Fitzgerald")
public_library.add_book("To Kill a Mockingbird", "Harper Lee")

public_library.display_books()

public_library.borrow_book("The Great Gatsby")
public_library.borrow_book("To Kill a Mockingbird")

public_library.display_books()

public_library.return_book("The Great Gatsby")

public_library.display_books()

```

```python
# 20. Describe the concept of method abstraction in Python and how it relates to polymorphism.

# Method Abstraction in Python:
    
# Definition:
# Method abstraction in Python involves creating abstract methods in a class, which are methods declared but 
# not implemented in the abstract base class.

# Abstract Base Classes (ABCs):
# Abstract base classes are used to define abstract methods in Python. The abc module provides the necessary tools
# for creating abstract classes and methods.

# @abstractmethod Decorator:
# The @abstractmethod decorator is used to declare abstract methods in an abstract base class. Subclasses must provide
# concrete implementations for these abstract methods.

#example:
from abc import ABC, abstractmethod

class MyAbstractClass(ABC):
    @abstractmethod
    def my_abstract_method(self):
        pass
    
# Enforcement of Structure:
# Abstract methods define a structure or interface that must be followed by the concrete subclasses. This ensures a 
# consistent interface across related classes.

# Relating to Polymorphism:
    
# Polymorphism in Python:
# Polymorphism is the ability of objects to take on multiple forms. In Python, it allows objects of different types to be
# treated as objects of a common type, enabling flexibility and code reuse.

# Achieving Polymorphism with Abstract Methods:

# Abstract methods play a key role in achieving polymorphism. Subclasses that inherit from an abstract base class
# can provide their own unique implementations for the abstract methods.

# Common Interface:
# Abstract methods create a common interface that multiple classes can adhere to. This shared interface allows objects
# of different classes to be used interchangeably if they implement the same abstract methods.

# Example:


from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def calculate_area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def calculate_area(self):
        return 3.14 * self.radius ** 2

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def calculate_area(self):
        return self.length * self.width

# Polymorphic Usage
circle_obj = Circle(5)
rectangle_obj = Rectangle(4, 6)

shapes = [circle_obj, rectangle_obj]

for shape in shapes:
    print(f"Area: {shape.calculate_area()}")
    
# In this example, both Circle and Rectangle classes adhere to the common interface defined by the abstract method 
# calculate_area(). The polymorphic behavior is demonstrated when objects of these classes are used interchangeably
# in a list and share a common method.

# In summary, method abstraction in Python, achieved through abstract methods, provides a common interface across related
# classes. This common interface, in turn, facilitates polymorphism by allowing objects of different classes to be treated 
# interchangeably when they implement the same abstract methods.
```

```python
# 1. Explain the concept of composition in Python and how it is used to build complex objects from simpler ones.

# Concept of Composition in Python:
    
# 1. Definition:
# Composition is a design principle in object-oriented programming where complex objects are created by combining 
# simpler objects as components.

# 2. Building Blocks:
# Instead of relying on inheritance, composition involves creating classes that contain instances of other classes
# as attributes. These instances are the building blocks or components.

# 3. Relationship:
# Composition establishes a "has-a" relationship between the containing class and its components. For example, a car 
# "has a" engine, wheels, and other components.

# 4. Code Reusability:
# Composition promotes code reusability by allowing the reuse of existing classes as components in new classes.

# 5. Flexibility:
# It provides greater flexibility than inheritance, as changes to the implementation of a component class do not affect
# the containing class as long as the interface remains consistent.

# 6. Encapsulation:
# Composition encapsulates the implementation details of the components within the containing class, reducing dependencies
# and promoting encapsulation.

class Engine:
    def start(self):
        return "Engine started"

class Wheels:
    def rotate(self):
        return "Wheels rotating"

class Car:
    def __init__(self):
        self.engine = Engine()  # Composition: Car has an Engine
        self.wheels = Wheels()  # Composition: Car has Wheels

    def drive(self):
        return f"{self.engine.start()} and {self.wheels.rotate()}"

# Creating a Car instance
my_car = Car()

# Using composition to build a complex object
result = my_car.drive()

# Output
print(result)  # Output: "Engine started and Wheels rotating"

# In this example:

# Engine and Wheels are simpler classes representing components.
# The Car class uses composition by containing instances of Engine and Wheels.
# The drive() method of Car utilizes the functionalities of its components.

```

```python
# 6. Create a Python class hierarchy for a music player system, using composition to represent playlists and songs.

class Song:
    def __init__(self, title, artist, duration):
        self.title = title
        self.artist = artist
        self.duration = duration

    def play(self):
        return f"Now playing: {self.title} by {self.artist}"

class Playlist:
    def __init__(self, name):
        self.name = name
        self.songs = []  # List to store Song objects

    def add_song(self, song):
        self.songs.append(song)

    def play_playlist(self):
        playlist_info = f"Playlist: {self.name}\n"
        for index, song in enumerate(self.songs, start=1):
            playlist_info += f"{index}. {song.title} - {song.artist} ({song.duration})\n"
        return playlist_info

class MusicPlayer:
    def __init__(self):
        self.playlists = []  # List to store Playlist objects

    def add_playlist(self, playlist):
        self.playlists.append(playlist)

    def play_all(self):
        player_info = "Music Player - All Playlists\n"
        for index, playlist in enumerate(self.playlists, start=1):
            player_info += f"{index}. {playlist.name}\n"
            player_info += playlist.play_playlist() + "\n"
        return player_info

# Example Usage
song1 = Song("Shape of You", "Ed Sheeran", "3:54")
song2 = Song("Dance Monkey", "Tones and I", "3:30")
song3 = Song("Blinding Lights", "The Weeknd", "3:22")

playlist1 = Playlist("Pop Hits")
playlist1.add_song(song1)
playlist1.add_song(song2)

playlist2 = Playlist("Top Tracks")
playlist2.add_song(song1)
playlist2.add_song(song3)

music_player = MusicPlayer()
music_player.add_playlist(playlist1)
music_player.add_playlist(playlist2)

# Play all playlists
result = music_player.play_all()
print(result)
```

```python
# 9. Describe the concept of "delegation" in composition and how it simplifies the design of complex systems.

# Concept of "Delegation" in Composition:
    
# Definition:
# "Delegation" in composition is a design pattern where one class (delegate) hands over responsibility for a
# specific behavior to another class (delegator).

# Mechanism:
# Achieved by creating an instance of another class within the delegator class and redirecting method calls or operations
# to the delegate class.

# Purpose:
# Simplifies the design by allowing classes to collaborate without inheritance, promoting flexibility and modularity.

# Key Components:
# Delegator Class: Contains an instance of the delegate class and delegates specific tasks to it.
# Delegate Class: Performs the delegated tasks and encapsulates related behavior.


# How Delegation Simplifies the Design of Complex Systems:
    
# Modularity:
# Breaks down complex systems into modular components with specific responsibilities, enhancing maintainability.

# Loose Coupling:
# Reduces dependencies between classes, leading to lower coupling and making the system more flexible.

# Code Reusability:
# Encourages code reusability by allowing the delegate class to be reused in various contexts.

# Dynamic Behavior:
# Permits dynamic changes to behavior during runtime, as the delegator can switch between different delegate implementations.
# Easy Extension:

# Facilitates easy extension by adding or modifying delegates without affecting the delegator class.

#example:
class InkjetPrinter:
    def print_document(self, document):
        return f"Inkjet printing: {document}"

class LaserPrinter:
    def print_document(self, document):
        return f"Laser printing: {document}"

class Printer:
    def __init__(self, printer_type):
        if printer_type == "inkjet":
            self.printer = InkjetPrinter()
        elif printer_type == "laser":
            self.printer = LaserPrinter()

    def print_document(self, document):
        return self.printer.print_document(document)

# Example Usage
inkjet_printer = Printer(printer_type="inkjet")
laser_printer = Printer(printer_type="laser")

print(inkjet_printer.print_document("Hello, Inkjet!"))
print(laser_printer.print_document("Hello, Laser!"))
```

```python
# 10. Create a Python class for a car, using composition to represent components like the engine, wheels, and transmission.

class Engine:
    def start(self):
        return "Engine started"

    def stop(self):
        return "Engine stopped"

class Wheels:
    def accelerate(self):
        return "Wheels accelerating"

    def brake(self):
        return "Wheels braking"

class Transmission:
    def shift_gear(self, gear):
        return f"Shifted to {gear} gear"

class Car:
    def __init__(self):
        self.engine = Engine()  # Composition: Car has an Engine
        self.wheels = Wheels()  # Composition: Car has Wheels
        self.transmission = Transmission()  # Composition: Car has Transmission

    def start(self):
        return self.engine.start()

    def drive(self):
        return f"{self.wheels.accelerate()} and {self.transmission.shift_gear('Drive')}"

    def stop(self):
        return f"{self.wheels.brake()} and {self.engine.stop()}"

# Example Usage
my_car = Car()

# Interacting with the car
print(my_car.start())    # Output: "Engine started"
print(my_car.drive())    # Output: "Wheels accelerating and Shifted to Drive gear"
print(my_car.stop())     # Output: "Wheels braking and Engine stopped"
```

```python
# 13. Discuss the challenges and drawbacks of composition, such as increased complexity and potential for tight coupling between objects.

# Challenges and Drawbacks of Composition:
    
# Increased Complexity:
# Challenge: As more components are composed, the overall structure can become complex, making it harder to
# understand and maintain.
# Drawback: Complexity may lead to difficulties in debugging, testing, and overall system comprehension.

# Potential for Tight Coupling:
# Challenge: Composition can result in tight coupling between classes, especially if there are strong dependencies 
# between components.
# Drawback: High coupling can lead to difficulties in modifying or extending the system without affecting other components.

# Dynamic Relationships:
# Challenge: Dynamically changing or swapping components during runtime can be challenging to manage.
# Drawback: Increases the risk of unexpected behaviors, making the system harder to predict and debug.

# Interface Proliferation:
# Challenge: Each component may introduce its own set of interfaces, potentially leading to a large number of methods
# in the composite class.
# Drawback: The interface may become cluttered, making it harder for users to identify the essential methods and properties.

# Inconsistency in Behavior:
# Challenge: If components have different behaviors or interfaces, composing them may lead to inconsistency.
# Drawback: Users of the composed class may face challenges in predicting how different components interact.

# Runtime Overhead:
# Challenge: The overhead of managing and coordinating interactions between components at runtime may affect performance.
# Drawback: In situations where performance is critical, composition may introduce unnecessary overhead.

# Dependency Management:
# Challenge: Managing dependencies between components, especially in larger systems, can be complex.
# Drawback: Incorrect dependency management can lead to issues such as circular dependencies or difficulties in resolving 
# dependencies during system initialization.

# Limited Code Reusability:
# Challenge: While composition promotes code reusability, it may not always be straightforward, especially if components 
# are tightly coupled.
# Drawback: Reusing components in different contexts might be challenging if they have specific dependencies or assumptions.

# Initialization Order Issues:
# Challenge: The order in which components are initialized can impact the behavior of the composed class.
# Drawback: Initialization order issues may lead to unexpected behaviors and make the system harder to reason about.

# Trade-off with Inheritance:
# Challenge: Choosing between composition and inheritance involves trade-offs. Composition may be more flexible but 
# requires more effort in managing interactions.
# Drawback: The chosen approach may not always align perfectly with the system's requirements, leading to compromises.
```

```python
# 17. Describe the concept of "aggregation" in composition and how it differs from simple composition.

# Concept of Aggregation
# Aggregation is a concept in which an object of one class can own or access another independent object of another class. 

# It represents Has-As relationship.
# It is a unidirectional association i.e. a one-way relationship. For example, a department can have students
# but vice versa is not possible and thus unidirectional in nature.
# In Aggregation, both the entries can survive individually which means ending one entity will not affect the other entity.

# Code to demonstrate Aggregation 

class Salary: 
	def __init__(self, pay, bonus): 
		self.pay = pay 
		self.bonus = bonus 

	def annual_salary(self): 
		return (self.pay*12)+self.bonus 

class EmployeeOne: 
	def __init__(self, name, age, sal): 
		self.name = name 
		self.age = age 
		self.agg_salary = sal # Aggregation 

	def total_sal(self): 
		return self.agg_salary.annual_salary() 

salary = Salary(10000, 1500) 
emp = EmployeeOne('Geek', 25, salary) 

print(emp.total_sal()) 

# we are not creating an object of the Salary class inside the EmployeeOne class, rather than that we are 
# creating an object of the Salary class outside and passing it as a parameter of EmployeeOne class

# Composition is defined by the PART-OF relationship which means that one object IS PART-OF ANOTHER OBJECT, 
# but Aggregation is defined by the HAS-A relationship which means that one object HAS-A RELATION with another object.

# In the case of Composition, the objects were interdependent on each other, but in Aggregation, they are Unidirectional
# means that as salary is a completely independent object we can only pass salary to the emp, not vice versa.
```

```python
# 18. Create a Python class for a house, using composition to represent rooms, furniture, and appliances.

class Furniture:
    def __init__(self, name):
        self.name = name

    def display_info(self):
        print(f"Furniture: {self.name}")

class Appliance:
    def __init__(self, name):
        self.name = name

    def display_info(self):
        print(f"Appliance: {self.name}")

class Room:
    def __init__(self, name):
        self.name = name
        self.furniture = []  # Composition: Room is composed of Furniture objects
        self.appliances = []  # Composition: Room is composed of Appliance objects

    def add_furniture(self, furniture):
        self.furniture.append(furniture)

    def add_appliance(self, appliance):
        self.appliances.append(appliance)

    def display_contents(self):
        print(f"\nRoom: {self.name}")
        print("Furniture:")
        for item in self.furniture:
            item.display_info()
        print("Appliances:")
        for item in self.appliances:
            item.display_info()

class House:
    def __init__(self):
        self.rooms = []  # Composition: House is composed of Room objects

    def add_room(self, room):
        self.rooms.append(room)

    def display_house_contents(self):
        print("House Contents:")
        for room in self.rooms:
            room.display_contents()

# Example Usage
living_room = Room(name="Living Room")
sofa = Furniture(name="Sofa")
tv = Appliance(name="Smart TV")
living_room.add_furniture(sofa)
living_room.add_appliance(tv)

kitchen = Room(name="Kitchen")
table = Furniture(name="Dining Table")
oven = Appliance(name="Oven")
kitchen.add_furniture(table)
kitchen.add_appliance(oven)

my_house = House()
my_house.add_room(living_room)
my_house.add_room(kitchen)

my_house.display_house_contents()

```

```python
# 19. How can you achieve flexibility in composed objects by allowing them to be replaced or modified dynamically at runtime?

# Achieving flexibility in composed objects, allowing them to be replaced or modified dynamically at runtime, can be 
# accomplished through various design patterns and principles. Here are some approaches:

# Strategy Pattern:
# Define a family of algorithms, encapsulate each algorithm, and make them interchangeable. This allows the client to 
# choose the appropriate algorithm dynamically.
# In the context of composed objects, different strategies can represent different behaviors or components that can be 
# switched at runtime.


# Dependency Injection:

# Instead of creating and managing dependencies within a class, inject dependencies from the outside. This allows the
# components to be replaced or modified without modifying the class.
# In Python, dependency injection can be achieved through constructor injection or method injection.

# Factory Pattern:
# Use a factory to create instances of components or objects. The factory can be configured to produce different 
# types of objects based on runtime conditions.
# This allows for the dynamic creation of objects without tightly coupling the client to concrete implementations.

# Abstract Factory Pattern:
# Define an interface for creating families of related or dependent objects without specifying their concrete classes.
# Clients can use the abstract factory to create sets of objects that work together, and the specific factory implementation
# can be swapped at runtime.

# Decorator Pattern:
# Attach additional responsibilities to an object dynamically. Decorators provide a flexible alternative to subclassing
# for extending functionality.
# In the context of composed objects, decorators can be dynamically added or removed to modify the behavior of the composed 
# object.

# Configuration or Settings:
# Store configuration or settings externally and load them dynamically at runtime. This allows for easy modification of 
# object behavior without code changes.
# Configuration can include specifying which components or strategies to use.

# Dynamic Binding and Polymorphism:
# Leverage dynamic binding and polymorphism in languages like Python. Design interfaces or abstract classes, and allow different implementations to be used interchangeably.
# Objects can be dynamically assigned to variables or passed as parameters based on runtime conditions.

```

```python
# 20. Create a Python class for a social media application, using composition to represent users, posts, and comments.

class Comment:
    def __init__(self, user, text):
        self.user = user
        self.text = text

    def display_info(self):
        print(f"{self.user.username}: {self.text}")

class Post:
    def __init__(self, user, content):
        self.user = user
        self.content = content
        self.comments = []  # Composition: Post is composed of Comment objects

    def add_comment(self, comment):
        self.comments.append(comment)

    def display_info(self):
        print(f"\nPost by {self.user.username}: {self.content}")
        print("Comments:")
        for comment in self.comments:
            comment.display_info()

class User:
    def __init__(self, username):
        self.username = username
        self.posts = []  # Composition: User is composed of Post objects

    def create_post(self, content):
        post = Post(user=self, content=content)
        self.posts.append(post)
        return post

# Example Usage
user1 = User(username="Alice")
user2 = User(username="Bob")

post1 = user1.create_post(content="Hello, social media world!")
comment1 = Comment(user=user2, text="Nice post, Alice!")
post1.add_comment(comment1)

post2 = user2.create_post(content="Enjoying the day!")
comment2 = Comment(user=user1, text="Great to hear, Bob!")
post2.add_comment(comment2)

# Display user posts and comments
for user in [user1, user2]:
    print(f"\n{user.username}'s Posts and Comments:")
    for post in user.posts:
        post.display_info()
```
